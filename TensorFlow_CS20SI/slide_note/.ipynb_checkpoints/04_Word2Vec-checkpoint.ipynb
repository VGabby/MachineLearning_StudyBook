```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import os
os.environ['TF_CPP_MIN_LOG_LEVEL']='2'
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.contrib.tensorboard.plugins import projector

from process_data import process_data
```

```python
VOCAB_SIZE = 50000
BATCH_SIZE = 128
EMBED_SIZE = 128 #dimension of the word embedding vectors ( Size of the hidden layer)
SKIP_WINDOW= 1 # the context window
NUM_SAMPLED= 64 # number of negative example to sample
LEARNING_RATE = 1.0
NUM_TRAIN_STEPS = 20000
SKIP_STEP = 2000 # how many steps to skip before reporting the loss 
```

```python
def word2vec(batch_gen):
    """ Build the graph for word2vec model and train it"""
    # Step 1 : define placeholder for input and output
    # center_word have to be int to work in embedding lookup
    #tf.reset_default_graph() # clean up graph
    with tf.name_scope('data'):
        center_words = tf.placeholder(tf.int32,shape=[BATCH_SIZE],name = "CenterWord") 
        target_words = tf.placeholder(tf.int32,shape=[BATCH_SIZE, 1],name = "TargetWord") # word to check
    # Step 2 : Define the weights, in word2vec, it's actually the weights that we card about 
    # vocab size x embed size
    # initialized to random uniform -1 to 1
    with tf.name_scope('embed'):
        embed_matrix = tf.Variable(tf.random_uniform([VOCAB_SIZE,EMBED_SIZE],-1.0,1.0),name = 'embed_matrix')
    # Step3 : Define the inference ( Compute the forward path of the graph)
    # get the embed of input words using tf.nn.embedding_lookup
    # embeded = tf.nn.embedding_lookup(embed_matrix,center_words, name = 'embed')
    with tf.name_scope('loss'):
        embed = tf.nn.embedding_lookup(embed_matrix, center_words, name = 'embed')
   # Step 4: Define loss function NCE
   # While NCE is cumbersome to implement in Python , TensorFlow already implemented it for us
   # tf.nn.nce_loss(weights, biases, labels, inputs, num_sampled, num_classes, ...)
   # nce_weight (vocab size x embed size), intialized to truncated_normal stddev=1.0 / (EMBED_SIZE ** 0.5)
   # bias: vocab size, initialized to 0
        nce_weight = tf.Variable(tf.truncated_normal([VOCAB_SIZE,EMBED_SIZE],stddev=1.0 / (EMBED_SIZE ** 0.5)),
                             name ='nce_weight')
                    

        nce_bias = tf.Variable(tf.zeros([VOCAB_SIZE]),name ="nce_bias")
    
        loss = tf.reduce_mean(tf.nn.nce_loss(weights=nce_weight,
                                          biases=nce_bias,
                                          labels=target_words,
                                          inputs=embed,
                                          num_sampled=NUM_SAMPLED,
                                          num_classes=VOCAB_SIZE), name='nce_loss')
    #Step 5 define optmizer
    optimizer = tf.train.GradientDescentOptimizer(learning_rate=LEARNING_RATE).minimize(loss)
    init = tf.global_variables_initializer()
    with tf.Session() as sess:
        # initialzie variables
        sess.run(init)
        total_loss = 0.0 # we use this to calculate the average loss in the las SKIP_STEP steps
        writer = tf.summary.FileWriter('./graphs/no_frills/',sess.graph)
        for index in range(NUM_TRAIN_STEPS):
            centers, targets = next(batch_gen)
            _,loss_batch = sess.run([optimizer,loss], feed_dict ={center_words: centers,target_words:targets})
            total_loss += loss_batch
            if (index + 1) % SKIP_STEP == 0:
                print('Average loss as step {}: {:5.1f}'.format(index, total_loss / SKIP_STEP))
                total_loss = 0.0
      
        
        ####################
        # code to visualize the embeddings. uncomment the below to visualize embedding    s
        # run "'tensorboard --logdir='processed'" to see the embeddings
        final_embed_matrix = sess.run(embed_matrix)
        # # it has to variable. constants don't work here. you can't reuse model.embed    _matrix
        embedding_var = tf.Variable(final_embed_matrix[:1000], name='embedding')
        sess.run(embedding_var.initializer)
        # Format: tensorflow/tensorboard/plugins/projector/projector_config.proto
        config = projector.ProjectorConfig()
        
        # # add embedding to the config file
        embedding = config.embeddings.add()
        embedding.tensor_name = embedding_var.name

        # # link this tensor to its metadata file, in this case the first 500 words of     vocab
        embedding.metadata_path = 'vocab_1000.tsv'
        summary_writer = tf.summary.FileWriter('processed')
        
        # # saves a configuration file that TensorBoard will read during startup.
        projector.visualize_embeddings(summary_writer, config)
        saver_embed = tf.train.Saver([embedding_var])
        saver_embed.save(sess, './processed/model3.ckpt', 1)
        
```

```python
def main():
    batch_gen = process_data(VOCAB_SIZE,BATCH_SIZE,SKIP_WINDOW)
    word2vec(batch_gen)
if __name__ == '__main__':
    main()
```
